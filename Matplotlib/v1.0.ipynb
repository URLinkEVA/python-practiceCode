# 绘制简单的折线图
使用平方数1，4，6，16，25来绘制这个图表

```python
import paddle
import matplotlib.pyplot as plt

squares = [1,4,9,16,25]
fig, ax = plt.subplots()
ax.plot(squares)

plt.show()
```

## 修改标签文字和线条粗细
通过一些定制来改善这个图表的可读性

```python
import paddle
import matplotlib.pyplot as plt

squares = [1,4,9,16,25]
fig, ax = plt.subplots()
ax.plot(squares, linewidth=3)

# 设置图表标题并给坐标轴加上标签
ax.set_title("Square number",fontsize=24)
ax.set_xlabel("Value",fontsize=14)
plt.ylabel("Square of value",fontsize=14)

# 设置刻度标记的大小
ax.tick_params(axis='both',labelsize=14)

plt.show()
```

linewidth决定了plot()绘制的线条粗细

fontsize指定图表中各种文字的大小

ax.set_xlabel为x轴设置标题，方法tick_params()设置刻度的样式，实参影响x轴与y轴上的刻度（axis='both'），字号设置为14

标签最好用英文，不然会不显示，要么使用plt.xlabel()方法

## 校正图形
折线图终点指向4.0的平方为25，下面来修复这个问题

向plot()提供一系列数时，改变默认第一个数据点对应x坐标值为0的行为，向plot（）同时提供输入值和输出值


```python
import paddle
import matplotlib.pyplot as plt

input_values = [1,2,3,4,5]
squares = [1,4,9,16,25]

fig,ax = plt.subplots()
ax.plot(input_values,squares,linewidth = 3)

# 设置图表标题并给坐标轴加上标签
```

下面来设置图表的样式，添加标签加标题

```python
import paddle
import matplotlib.pyplot as plt

plt.style.use('seaborn')
fig,ax = plt.subplots()
ax.scatter(2,4,s=200)

# 设置图表标题标签
ax.set_title("Square number",fontsize=24)
ax.set_xlabel("Value",fontsize=14)
ax.set_ylabel("Square of value",fontsize=14)

# 设置刻度标记的大小
ax.tick_params(axis='both',which='major',labelsize=14)

plt.show()
```

## 自动计算数据
绘制1000个点的数据

```python
import paddle
import matplotlib.pyplot as plt

x_values = range(1,1001)
y_values = [x**2 for x in x_values]

plt.style.use('seaborn')
fig,ax = plt.subplots()
ax.scatter(x_values,y_values,s = 10)

# 设置每个坐标轴的取值范围
ax.axis([0,1100,0,1100000])
```

## 自定义颜色
修改数据点颜色，向scatter()传递参数c,并将其设置为要使用的颜色名称

```python
ax.scatter(x_values,y_values,c='red',s=10)
```

还可以使用 RGB 颜色模式自定义颜色，传递参数c，并设置为一个元组，三个0~1的小数值，分别表示红绿蓝颜色的分量，例如下面由淡绿色点组成的散点图：

```python
ax.scatter(x_values,y_values,c=(0,0.8,0),s=10)
```

值越接近0，指定的颜色越深；值越接近1，指定的颜色越浅

## 使用颜色映射
从起始颜色渐变到结束颜色，在可视化中，颜色映射用于突出数据的规律

模块pyplot内置了一组颜色映射，下面演示如何根据每个点的y值来设置其颜色

```python
import paddle
import matplotlib.pyplot as plt
fig,ax = plt.subplots()

x_values = range(1,1001)
y_values = [x**2 for x in x_values]

ax.scatter(x_values,y_values,c=y_values,cmap=plt.cm.Blues,s=10)

#然后设置标题和标签
```

将参数c设置成了一个y值列表，并使用参数cmap告诉pyplot使用哪个颜色映射，将y值较小的点显示为浅蓝色，y值较大的点显示为深蓝色
